```python
# 加载数据
cifar10.maybe_download_and_extract()
class_names=cifar10.load_class_names()
print(class_names)
# 加载训练数据和测试数据
images_train,cls_train,labels_train=cifar10.load_training_data()
images_test,cls_test,labels_test=cifar10.load_test_data()
```

```python
# 数据可视化
def plot_images(images, cls_true, cls_pred=None, smooth=True):
    assert len(images) == len(cls_true)

    # 划分子窗口
    fig, axes = plt.subplots(3, 3)

    # 子窗口间隔
    if cls_pred is None:
        hspace = 0.3
    else:
        hspace = 0.6
    fig.subplots_adjust(hspace=hspace, wspace=0.3)

    # 填充方式
    if smooth:
        interpolation = 'spline16'
    else:
        interpolation = 'nearest'

    for i, ax in enumerate(axes.flat):
        if i < len(images):
            # 显示图片
            ax.imshow(images[i],
                      interpolation=interpolation)

            # 图片标签
            cls_true_name = class_names[cls_true[i]]

            # 显示图片和图片标签
            if cls_pred is None:
                xlabel = "True: {0}".format(cls_true_name)
            else:
                # 预测结果
                cls_pred_name = class_names[cls_pred[i]]

                xlabel = "True: {0}\nPred: {1}".format(cls_true_name, cls_pred_name)

            # 显示x轴标题
            ax.set_xlabel(xlabel)

        # 清楚刻度
        ax.set_xticks([])
        ax.set_yticks([])
    plt.show()
```

```python
# 图片
images = images_test[0:9]

# 标签
cls_true = cls_test[0:9]
plot_images(images=images, cls_true=cls_true, smooth=False)
```

<h1>Tensorflow构建神经网络分类器</h1>

<p style='color="green";'>首先找到transfer_values的数组长度</p>

```python
transfer_len=model.transfer_len
```

```python
print(transfer_len)
```

```python
x=tf.placeholder(tf.float32,shape=[None,transfer_len],name='x')
```

```python
y_true=tf.placeholder(tf.float32,shape=[None,num_classes],name='y_true')
y_true_cls = tf.argmax(y_true, dimension=1)
```

```python
layer1=tf.layers.dense(inputs=x,units=1024,activation=tf.nn.relu)
layer2=tf.layers.dense(inputs=layer1,units=10)
y_pred=tf.nn.softmax(layer2,)
```

```python
y_pred_cls = tf.argmax(y_pred, dimension=1)         # 真实类别
```

```python
'''
代价函数
概率交叉熵
'''
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits=y_pred,
                                                        labels=y_true)
cost = tf.reduce_mean(cross_entropy)
```

```python
'''
优化方法
'''
optimizer = tf.train.AdamOptimizer(learning_rate=1e-4).minimize(cost)
```

```python
'''
性能度量
'''

correct_prediction = tf.equal(y_pred_cls, y_true_cls)
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

<h1>运行tensorflow</h1>

<h3>创建tensorflow会话</h3>
<p style='color:blue;'>一旦创建了TensorFlow图，我们需要创建一个TensorFlow会话，用来运行图。<p>

```python
session=tf.Session()
```

<h3>初始化变量</h3>
<p style='color:blue;'>我们需要在开始优化weights和biases变量之前对它们进行初始化。<p>

```python
global_step = tf.Variable(initial_value=0,name='global_step', trainable=False)
session.run(tf.global_variables_initializer())
```

<h3>获取随机训练batch的帮助函数</h3>
<p style='color:blue;'>训练集中有50,000张图像（以及保存transfer-values的数组）。
用这些图像（transfer-vlues）计算模型的梯度会花很多时间。
因此，我们在优化器的每次迭代里只用到了一小部分的图像（transfer-vlues）。</p>
<p style='color:blue;'>
</p>

```python
train_batch_size = 64
def random_batch():
    # 获取训练数据的数量
    num_images = len(transfer_values_train)

    # 创建随机训练的下标.
    idx = np.random.choice(num_images,
                           size=train_batch_size,
                           replace=False)

    # 获取随机下标对应的训练数据与标签
    x_batch = transfer_values_train[idx]
    y_batch = labels_train[idx]

    return x_batch, y_batch
```

```python
def optimize(num_iterations):
    # 开始训练的事假
    start_time = time.time()

    for i in range(num_iterations):
        # 产生随机批量数据
        x_batch, y_true_batch = random_batch()

        # 指定占位符实参
        feed_dict_train = {x: x_batch,
                           y_true: y_true_batch}

        # 运行优化运算图
        i_global, _ = session.run([global_step, optimizer],
                                  feed_dict=feed_dict_train)

        #每一百次输出一次性能评估的结果
        if (i % 100 == 0) or (i == num_iterations - 1):
            # 计算准确率
            batch_acc = session.run(accuracy,
                                    feed_dict=feed_dict_train)

            # 打印状态
            msg = "Global Step: {0:>6}, Training Batch Accuracy: {1:>6.1%}"
            print(msg.format(i, batch_acc))

    # 结束时间
    end_time = time.time()

    # 计算迭代时间
    time_dif = end_time - start_time

    # 打印输出
    print("Time usage: " + str(timedelta(seconds=int(round(time_dif)))))
```

```python
def plot_example_errors(cls_pred, correct):
    # 所有测试数据的测试结果

    # correct ：表示在测试样本预测是否正确
    incorrect = (correct == False)
    
    #获取预测错误的样本
    images = images_test[incorrect]
    
    # 获取预测错误的预测结果
    cls_pred = cls_pred[incorrect]

    # 获取预测错误的样本标签
    cls_true = cls_test[incorrect]

    n = min(9, len(images))
    
    # 绘制图片
    plot_images(images=images[0:n],
                cls_true=cls_true[0:n],
                cls_pred=cls_pred[0:n])
```

```python
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(cls_pred):
    cm = confusion_matrix(y_true=cls_test,  # 测试样本标签
                          y_pred=cls_pred)  # 预测类别

    # 打印混淆矩阵
    for i in range(num_classes):
        # 连接类别名称
        class_name = "({}) {}".format(i, class_names[i])
        print(cm[i, :], class_name)


    class_numbers = [" ({0})".format(i) for i in range(num_classes)]
    print("".join(class_numbers))
```

```python
def predict_cls_test():
    return predict_cls(transfer_values = transfer_values_test,
                       labels = labels_test,
                       cls_true = cls_test)
```

```python
def classification_accuracy(correct):
    return correct.mean(), correct.sum()
```

```python
def print_test_accuracy(show_example_errors=False,
                        show_confusion_matrix=False):
    correct, cls_pred = predict_cls_test()
    
    acc, num_correct = classification_accuracy(correct)
    num_images = len(correct)

    msg = "Accuracy on Test-Set: {0:.1%} ({1} / {2})"
    print(msg.format(acc, num_correct, num_images))

    if show_example_errors:
        print("Example errors:")
        plot_example_errors(cls_pred=cls_pred, correct=correct)

    if show_confusion_matrix:
        print("Confusion Matrix:")
        plot_confusion_matrix(cls_pred=cls_pred)
```
